```python
tokenizer = RegexpTokenizer(r'[A-z\'`a-z]+')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
spellchecker = SpellChecker()

def tokenize(w):
    tokens = tokenizer.tokenize(w)
    for token in tokens:
        if token.lower() in contractions:
            tokens.remove(token)
            for newToken in tokenizer.tokenize(contractions[token.lower()]):
                tokens.append(newToken)
    tokens = spellchecker.known([token.lower() for token in tokens if token.lower() not in stop_words])
    return [lemmatizer.lemmatize(token) for token in tokens]
```

```python
df['comment_text'] = df['comment_text'].apply(tokenize)
```
